```python
# set cutoff minimum rating

min_rating = data['Total_Rating'].quantile(0.50)
print(min_rating)
```

```python
# set cutoff minimum users rated

min_total_user_rated = data['Users_Rated'].quantile(0.30)
print(min_total_user_rated)
```

```python
#to check average how many CPU have rating min_rating or above

avg_ds_min_rating =  data.copy().loc[data['Total_Rating'] >= min_rating]
avg_ds_min_rating.shape
```

```python
#to check average how many CPU have total users rated min_total_user_rated or above

avg_ds_users_rated =  data.copy().loc[data['Users_Rated'] >= min_total_user_rated]
avg_ds_users_rated.shape
```

```python
# formula to find and calculate score for best rated machines

def weighted_rating(x, min_total_user_rated=min_total_user_rated, avg_rating =  avg_rating):
    users = x['Users_Rated'] 
    avg_total_rating = x['Total_Rating']
    return (users/(users+min_total_user_rated) * avg_total_rating) + (min_total_user_rated/(min_total_user_rated+users) * avg_rating)
```

```python
# store the values in score column

avg_ds_users_rated['score'] = avg_ds_users_rated.apply(weighted_rating, axis=1)
```

```python
# sorting the data/results 

avg_ds_users_rated =  avg_ds_users_rated.sort_values('score', ascending=True)
```
